```python
#tableau crois√©s dynamiques
#Ces tableaux, encore appel√©s tables de pivots (ou pivot table), permettent de synth√©tiser
#les donn√©es contenues dans un DataFrame. Essayons de voir cela par l'exemple.

titanic.pivot_table("survived", index="sex", columns="class", aggfunc="sum")
```

```python
#tableau crois√©s dynamiques
titanic.pivot_table("survived", index="sex", columns="class")
#sans le sum donne la moyenne
```

```python
#La fonction pivot_table est tr√®s puissante, et permet m√™me de faire des agr√©gations √† plusieurs niveaux.
#Par exemple, nous pouvons voir l'√¢ge des survivants comme une dimension suppl√©mentaires.
#Le nombre exact d'ann√©es nous int√©ressant peu, nous regrouperons les √¢ge en deux cat√©gories,
#gr√¢ce √† la fonction  cut.

titanic.dropna(inplace=True)
#modifie directement le DataFrame

age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')

#on a 2 niveau d'indexation √† gauche
```
