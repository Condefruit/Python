```python
#tableau croisés dynamiques
#Ces tableaux, encore appelés tables de pivots (ou pivot table), permettent de synthétiser
#les données contenues dans un DataFrame. Essayons de voir cela par l'exemple.

titanic.pivot_table("survived", index="sex", columns="class", aggfunc="sum")
```

```python
#tableau croisés dynamiques
titanic.pivot_table("survived", index="sex", columns="class")
#sans le sum donne la moyenne
```

```python
#La fonction pivot_table est très puissante, et permet même de faire des agrégations à plusieurs niveaux.
#Par exemple, nous pouvons voir l'âge des survivants comme une dimension supplémentaires.
#Le nombre exact d'années nous intéressant peu, nous regrouperons les âge en deux catégories,
#grâce à la fonction  cut.

titanic.dropna(inplace=True)
#modifie directement le DataFrame

age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')

#on a 2 niveau d'indexation à gauche
```
